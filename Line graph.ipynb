```python
from matplotlib import pyplot as plt
import numpy as np
```

# SIMPLE LINE GRAPH

```python
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

fig,ax = plt.subplots()
# fig = plt.figure()
# ax = fig.add_subplot()

ax.plot(years, chemical_exports)

plt.title('line graph')

ax.set_xlabel('years')
ax.set_ylabel('chemical_exports')

x_vals = np.arange(2010,2022,2)
y_vals = np.linspace(0.80,0.98,10)
ax.set_xticks(x_vals)
ax.set_yticks(y_vals)

plt.show()
```

```python
medicine_exports = [0.791, 0.818, 0.832, 0.816, 0.840, 0.833, 0.835, 0.838, 0.842, 0.910, 0.930, 0.940]
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

fig,ax = plt.subplots()

ax.plot(years, chemical_exports, label = 'chemical')
ax.plot(years, medicine_exports, label = 'medicine')

plt.title('line graph')

ax.set_xlabel('years')
ax.set_ylabel('chemical_exports')

x_vals = np.arange(2010,2022,2)
y_vals = np.linspace(0.80,0.98,10)
ax.set_xticks(x_vals)
ax.set_yticks(y_vals)


plt.legend(loc = 'best')

plt.show()
```

```python
medicine_exports = [0.791, 0.818, 0.832, 0.816, 0.840, 0.833, 0.835, 0.838, 0.842, 0.910, 0.930, 0.940]
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

# fig,ax = plt.subplots()
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot()


ax.plot(years, chemical_exports, label = 'chemical', marker = 'x', c = 'black', ls = '-', lw = 2)
ax.plot(years, medicine_exports, label = 'medicine', marker = 'o', c = 'green', ls = ':', lw = 2)

plt.title('line graph')

ax.set_xlabel('years')
ax.set_ylabel('chemical_exports')

x_vals = np.arange(2010,2022,2)
y_vals = np.linspace(0.80,0.98,10)
ax.set_xticks(x_vals)
ax.set_yticks(y_vals)


plt.annotate("Peak of Chemical Exports", xy=(2016,0.95))


plt.legend(loc = 'best')

plt.grid()

plt.show()
```

```python
from matplotlib import pyplot as plt
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
medicine_exports = [0.791, 0.818, 0.832, 0.816, 0.840, 0.833, 0.835, 0.838, 0.842, 0.910, 0.930, 0.940]

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)

ax1.plot(years, chemical_exports, label='Chemicals', marker='o', c='b', ls='-', lw=2) 
ax1.set_title("LCI Chemical exports in last 12 years")
ax1.set_xlabel("Years")
ax1.set_ylabel("Amount (Million Rs.)")

ax2.plot(years, medicine_exports, label='Medicines', marker='x', c='r', ls=':', lw=2) 
ax2.set_title("LCI Medicine exports in last 12 years")
ax2.set_xlabel("Years")
ax2.set_ylabel("Amount (Million Rs.)")

plt.tight_layout()
plt.show()
```
