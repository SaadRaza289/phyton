# Scatter plot

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
l1 = np.random.randint(0,100,20)
l2 = np.random.randint(0,100,20)

fig, ax = plt.subplots()

ax.scatter(x = l1, y = l2)

plt.show()
```

```python
l1 = np.random.randint(0,100,20)
l2 = np.random.randint(0,100,20)

fig, ax = plt.subplots()

ax.scatter(x = l1, y = l2, s = 200, marker = 'o', edgecolors= 'black', lw = 1)


ax.set_xlabel('list1')
ax.set_ylabel('list2')

plt.title('scatter plot')

# cbar = plt.colorbar()

plt.grid()

plt.show()
```
